```python
from google.colab import drive
drive.mount('/content/drive')

import import_ipynb
from utils.dataset import CVRDataset
from model.model import CVRModel

from torch.utils.data import DataLoader
import torch.nn.functional as F
import torch

BATCH = 20000
EPOCHS = 200

def train(mode, model):
    train_data=CVRDataset(mode)
    train_loader=DataLoader(train_data, batch_size=BATCH, shuffle=True)
    print("Train Data Loading Success! Size is {}.".format(len(train_loader.dataset)))
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
    model.train()

    for epoch in range(EPOCHS):
        for batch_idx, samples in enumerate(train_loader):
            x_train, y_train=samples  ## x_train size : [bs, 22]
            logits = model(x_train)
            loss=F.binary_cross_entropy(logits.squeeze(), y_train)
            loss.retain_grad()
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            print('Epoch {:4d}/{} Batch {}/{} Loss: {:.6f}'.format(epoch+1, EPOCHS, batch_idx+1,len(train_loader), loss.item()))
    print("Train Finished")
  
def test(mode, model):
    test_data=CVRDataset(mode)
    test_loader=DataLoader(test_data, batch_size=BATCH, shuffle=True)
    print("Test Data Loading Success! Size is {}.".format(len(test_loader.dataset)))
    model.eval()
    answer_cnt=0

    for batch_idx, samples in enumerate(test_loader):
        x_test, y_test=samples  ## x_train size : [bs, 21]
        logits = model(x_test)
        logits=logits.view(-1)
        answer_cnt+=torch.sum(torch.round(logits)==y_test)
    print("Test Accuracy : {}".format(answer_cnt/len(test_data)*100))
    print("Test Finished")


model = CVRModel(21,1)
train("Train",model)
test("Test", model)
torch.save(model.state_dict(),"./weight/train_model.dat")


```
